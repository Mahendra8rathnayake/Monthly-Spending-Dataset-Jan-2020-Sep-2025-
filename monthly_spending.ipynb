# Data Cleaning

## Replace $ with (₹) in column names

```python
df.columns = df.columns.str.lower().str.replace(r"\(₹\)", "$", regex=True)
```

```python
df.describe()
```

## Rename the columns without $ 

```python

df.columns = (
    df.columns
    .str.lower()
    .str.replace("$", "", regex=False)          # remove dollar sign
    .str.replace(" ", "_", regex=False)         # replace spaces
)
```

```python
df
```

## Convert month column to datetime

```python
df['month'] = pd.to_datetime(df['month'])
df = df.sort_values('month')
```

```python
df
```

## Sum only numeric columns

```python
df.select_dtypes(include='number').sum().sort_values(ascending=False)
```

# Financial & Behavioral Metrics Calculation:

```python
# Monthly Savings Rate (%)

# How much of the income was saved.

df['savings_rate_'] = (df['savings_'] / df['income_']) * 100
df['savings_rate_'].mean()
```

```python
# Spending-to-Income Ratio

# How much of income was spent.

df['spending_ratio_'] = df['total_expenditure_'] / df['income_']
df['spending_ratio_'].mean()
```

```python
# Investment Rate (%)

# What portion of income is invested.

df['investment_rate_'] = (df['investments_'] / df['income_']) * 100
df['investment_rate_'].mean()
```

```python
# Monthly Net Savings

# Income - Total Expenditure.

df['net_savings_'] = df['income_'] - df['total_expenditure_']
df['net_savings_'].mean()
```

```python

# Year-over-Year Spending Growth

# Shows financial improvement or decline.

df['year'] = df['month'].dt.year
df.groupby('year')['total_expenditure_'].sum().pct_change()
```

```python
# Category-wise Spending Growth

# Identifies which categories increased the most.

category_growth = df.groupby('year')[df.columns[1:-2]].sum().pct_change()
category_growth
```

```python
# Lowest Spending Month
df.loc[df['total_expenditure_'].idxmin()]
```

```python
# Highest Saving Month
df.loc[df['savings_'].idxmax()]
```

```python

# Correlation Between Income & Spending
df[['income_', 'total_expenditure_']].corr()
```

```python
# Categories That Correlate Most With Total Expenditure
df.corr(numeric_only=True)['total_expenditure_'].sort_values(ascending=False)
```

```python

# Average Monthly Spending by Category
df.mean(numeric_only=True).sort_values(ascending=False)
```

```python
# Median Monthly Spending
df.median(numeric_only=True)
```

```python
# Standard Deviation (Spending Stability)
df.std(numeric_only=True)
```

```python
# Coefficient of Variation (Financial Stability Metric)

# Lower = more stable.

(df.std(numeric_only=True) / df.mean(numeric_only=True)).sort_values()
```

```python

# Months Overspending (Expenditure > Income)
df[df['total_expenditure_'] > df['income_']]
```

```python
# % of Months Overspent
overspent_rate = (df['total_expenditure_'] > df['income_']).mean() * 100
overspent_rate
```
